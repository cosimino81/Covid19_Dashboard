```python
import pandas as pd
import pycountry_convert as pc
```

```python
# define function to convert country code to continent name
def country_to_continent(country_code):
    if country_code == "JPG11668":
        country_code = "JP"
    # obsolete codes   
    elif country_code == "EL":
        country_code = "GR"
        
    elif country_code == "VA":
        country_code = "IT"
        
    elif country_code == "SX":
        country_code = "HT"
    
    elif country_code == "TL":
        country_code = "JP"
        
    elif country_code == "UK":
        country_code = "GB"
    
    elif country_code == "":
        country_code = "NA"
        
    country_continent_code = pc.country_alpha2_to_continent_code(country_code)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name

# Example
country_code = 'NA'

print(country_to_continent(country_code))
```

```python
df = pd.read_csv("WHO-COVID-19-global-data.csv")
df
```

```python
df = df.dropna(axis=0)
df.shape
```

```python
df.Country.unique()
```

```python
df
```

```python
df.shape
```

```python
df["Continent"] = df["Country"].apply(lambda x: country_to_continent(x))
```

```python
df
```

```python
df.Continent.unique()
```

```python
df = df.groupby(["day","Country","Country Name" ,"Continent"]).sum().reset_index()
df
```

```python
import plotly.express as px

#df = px.data.gapminder().query("continent=='Oceania'")

fig = px.line(df, x="day", y="Deaths", color='Continent')
fig.show()
```

```python
import plotly.express as px
#df = px.data.gapminder()
fig = px.scatter_geo(df, locations="Country", color="Continent",
                     hover_name="Country Name", size="Cumulative Deaths",
                     #animation_frame="Period",
                     projection="natural earth")


# Set theme, margin, and annotation in layout
fig.update_layout(
    template="plotly_dark",
    margin=dict(r=10, t=25, b=40, l=60),
    annotations=[
        dict(
            text="Source: UE Open Data",
            showarrow=False,
            xref="paper",
            yref="paper",
            x=0,
            y=0)
    ]
)
fig.show()
```
